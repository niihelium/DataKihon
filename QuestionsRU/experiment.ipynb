сделать?

классификаторами, т.е. они строят сложную разделяющую поверхность. Почему тогда нам интуитивно проще понимать дерево решение, а не глубокую нейросеть? 
- K-fold cross validation плохо подходит для работы с временными рядами. Почему?
Как это исправить?

тестовую хорошо работает в случае задач регрессии. Что может пойти не так при использовании этого же метода в задачи классификации? Что с этим можно сделать?
